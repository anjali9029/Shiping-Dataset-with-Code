```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from CSV file
data = pd.read_csv('/content/shippppppppp1.csv')

# Display basic information about the dataset
print(data.info())

# Summary statistics of numerical columns
print(data.describe())

# Count of Ship Types
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Ship Type')
plt.xlabel('Ship Type')
plt.ylabel('Count')
plt.title('Count of Ship Types')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Count of Pollutants
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Pollutant')
plt.xlabel('Pollutant')
plt.ylabel('Count')
plt.title('Count of Pollutants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Time-based analysis using Date
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

# Count of records over the years
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of Records over the Years')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Count of records per month
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Count of Records per Month')
plt.tight_layout()
plt.show()

# Correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('/content/shippppppppp1.csv')

# Filter rows where "Ship Type" is "Defence"
defence_data = data[data['Ship Type'] == 'Defence']

# Group by "Pollutant" and calculate the mean "Estimated Litres"
pollutant_mean_litres = defence_data.groupby('Pollutant')['Estimated Litres'].mean()

# Create a line graph
plt.figure(figsize=(7, 4))
pollutant_mean_litres.plot(marker='o')
plt.xlabel('Pollutant')
plt.ylabel('Mean Estimated Litres')
plt.title('Mean Estimated Litres for Defence Ships by Pollutant')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load CSV data into a pandas DataFrame
data = pd.read_csv('/content/commercial1.csv')

# Assuming 'Date' and 'Estimated Liters' are the column names
data['Date'] = pd.to_datetime(data['Date'])

# Extract data for linear regression
X = data['Date'].astype(int).values.reshape(-1, 1)  # Convert dates to integer values
y = data['Estimated Litres'].values

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict values using the model
y_pred = model.predict(X)

# Calculate metrics
rmse = mean_squared_error(y, y_pred, squared=False)
r2 = r2_score(y, y_pred)

# Display metrics
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (Coefficient of Determination): {r2:.2f}")

# Create a line plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Date'], data['Estimated Litres'], color='blue', label='Actual Data')
plt.plot(data['Date'], y_pred, color='red', label='Linear Regression')
plt.xlabel('Date')
plt.ylabel('Estimated Liters')
plt.title('Linear Regression: Estimated Liters Over Time')
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load CSV data into a pandas DataFrame
data = pd.read_csv('/content/merged_dataset.csv')

# Assuming 'Date' and 'Estimated Liters' are the column names
data['Date'] = pd.to_datetime(data['Date'])

# Extract data for linear regression
X = data['Date'].astype(int).values.reshape(-1, 1)  # Convert dates to integer values
y = data['Estimated Litres'].values

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict values using the model
y_pred = model.predict(X)

# Calculate metrics
rmse = mean_squared_error(y, y_pred, squared=False)
r2 = r2_score(y, y_pred)

# Display metrics
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (Coefficient of Determination): {r2:.2f}")

# Create a line plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Date'], data['Estimated Litres'], color='blue', label='Actual Data')
plt.plot(data['Date'], y_pred, color='red', label='Linear Regression')
plt.xlabel('Date')
plt.ylabel('Estimated Liters')
plt.title('Linear Regression: Recreational Ship Estimated Liters Over Time')
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
